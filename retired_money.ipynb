```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fontManger

zhfont=fontManger.FontProperties(fname='NotoSansTC-VariableFont_wght.ttf')
noYear=16
x=np.arange(1,noYear,1)
n=len(x)
yTotal=np.zeros(n)
yYear=np.zeros(n)

baseMoney=8*12 # 
rateSave=0.03  # 
rateJob=0.04   #  
yTotal[0]=baseMoney  # 
yYear[0]=baseMoney   #
for i in range(n-1):
    yYear[i+1]=yYear[i]*(1+rateJob)
    yTotal[i+1]=(yTotal[i]+yYear[i+1])*(1+rateSave)
#-------
fig, ax1 = plt.subplots()
# plt.title('Numbers of blocks NN opened per Game')
plt.xlabel('', fontproperties=zhfont)
ax2 = ax1.twinx()

ax1.set_ylabel('()',fontproperties=zhfont)

# ax1.plot(range(len(data)), data, color='tab:blue', alpha=0.75)
# ax1.tick_params(axis='y', labelcolor='tab:blue')

ax1.plot(x,yTotal,'b-',label='')
ax1.axhline(y=1800,linestyle='dotted',color='red',label='1800')
ax1.axhline(y=2400,linestyle='dotted',color='brown',label='2400')
ax1.axvline(x=12,linestyle='dotted',color='black')
ax1.grid()

ax2.set_ylabel('()',fontproperties=zhfont, color='black')

ax2.plot(x,yYear,'g--')
ax2.tick_params(axis='y', labelcolor='green')
ax2.set_ylim([90,250])
# plt.ylabel('()',fontproperties=zhfont)
plt.xticks(np.arange(1, noYear, 1.0))
ax1.legend(loc='upper left',prop=zhfont)
# plt.grid()
plt.show()
```
